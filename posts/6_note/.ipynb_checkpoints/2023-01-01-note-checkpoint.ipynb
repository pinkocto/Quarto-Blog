제품별 가격/판매량이 포함된 시계열 데이터 $\to$ 미래 3주간 제품별 판매량 예측

## 1. 소비재

### 1-1. 소비재의 정의

소비재는 경제학에서 산업재에 상응하는 말이지만, 마트에서 살 수 있는 제품이라 생각하면 편리.

예시) 샴푸, 세제, 옷, 기저귀, 물티슈

`-` 소비재의 특징

- 잦은 / 반복적인 소비
- 저관여
- 낮은 가격
- 짧은 유통기한과 빠른 소비
- 접근성과 인지도가 판매에 직접적인 영향
- 습관적 구매 비중이 높음.

### 1-2. 소비재 특징


#### 1). 제품 파편화

소비재 제품은 대체로 브랜드 기준으로 구분되지만, 출시 시점 이후 제품의 세분화/파편화가 빠르게 진행.

`-` **브랜드 라인으로 파편화**

초창기 탈모 전용 샴푸로 출시되던 닥터그루트 $\to$ 세분화된 니즈 충족을 위해 청소년 특이취용 / 향이 추가된 / 여름 전용 쿨링 샴푸 등이 운영

`-` **패키징/용량의 파편화**

Fiji 시트형 세제 $\to$ 번들 / 향, 용량변경

#### 2). 제품군/제품 내에서의 상관관계

소비자의 선호도, 사용 습관에 의해 제품별 판매량은 일정한 비율이 유지되는 경향.

- 함께 사용하는 제품군일 경우 비율이 유지되는 경향. (예. 샴푸와 린스, 트리트먼트, 세탁세제와 섬유유연제)
- 특정 제품이 default 옵션으로 인식되는 경우, 혹은 특징(향)의 선호도가 존재하는 경우 (예. 대용량/보통 용량/ 소용량 제품 간, 꽃 향기 제품과 일반 코튼 향기 제품)

### 1-3. 소비재 시장 내 최근 트랜드

#### 1. 제품의 성분/패키징에 대한 관여도 증가
- 심미적 디자인 외적으로 친환경적 제품인지, 특정 성분(파라벤, 실리콘 등)이 없는지를 확인하는 소비자 증가.

#### 2. 콜라보를 통한 인지도 증가 노력
- 캐릭터가 추가된 콜라보 제품으로 소장욕을 증가시키거나, 유튜버와 협업하여 타겟 소비자에 소구.

## 2. 수요예측

1. 수요예측을 왜 하는지?
2. 소비재 수요예측의 어려운 점
3. 피처를 어떻게 찾을 것인지?

### 2-1. 수요예측이 필요한 이유

제조사는 제조, 원료 구매 계획을 세워야하며 Lead Time을 고려하여 계획해야함.
- 채찍효과: 수요에 변동이 생산계획 및 구매 계획 변동이 점점커짐
- 과소예측 $\to$ 판매 실기 (기회비용)
- 과다예측 $\to$ 재고 관리 비용, 처분 비용 발생

### 2-2. 소비재 수요예측의 특징: spike성 데이터

Outlier적인 판매량이 자주 발생, 제품별 패턴 상이
- 발생 이유는 할인행사, 입소문, 관고 등 다양.

### 소비재 특징: 이벤트에 따른 수요 변동


`-` 1. 주기적이벤트

**전역적인 이벤트**

- 국내 설, 추석 기간 생활용품/가공식품 선물세트 제품의 판매량 증가.
- 코리아세일페스타는 매년 11월 브랜드별 할인행사 진행.

**제조사/유통사별 정기 할인행사**

- 제조사/유통사별로 정기적(분기/연령) 대대적인 할인행사 진행 $\to$ 수요 spike
- 엘지생활건강 레드워크, 올리브영 올영세일, 이마트 창립기념일

`-`  2. 비주기적 이벤트

비정기적으로 예측하기 어려운 이벤트

**Word of mouth**
- 특정 연예인의 SNS/인터뷰 등에서 브랜드 언급이 있으면 관심도 폭증
- 최근 몇 년간 유튜버 영향도 급증

**재난/사건사고**

- 예측하기 어렵지만 임팩트가 큰 사건 발생
- 예) 코로나 전염위험에 의한 수요 변동, 특정 국가 불매운동 등.

### 2-3. 피처를 어떻게 찾을 것인가

- 소비재의 특징을 위주로 피처 추출
    - 제품적 특징: 브랜드 인지도/충성도, 구매접근성, 가격 포지셔닝, 가격 민감도, 본품 여부(리필), 대용량 여부
    - 이벤트: 명절인지, 할인기간인지, 광고를 진행중인지

- 제품군별 특징을 고려해야함: <font color='red'>생활필수품 적인 제품군은 전체 수요가 급변동하지 않음.</font>

- 판매량 데이터는 수요와 공급이 만난 구간이다. 공급이 부족했다면, 이상치 수준의 가격변동이라면 판매량에 영향.

> *판매량이 0 $\to$ 제품마다 다른의미* $\to$ 말그대로 판매가 $0$일수도 있고 공급이 부족해서 판매량이 $0$이었을 수도 있다.
