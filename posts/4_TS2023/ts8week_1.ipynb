비정상 시계열의 대표적인 특성: 
1. 시계열의 수준이 시간에 따라 다르다. (수준=평균) 
2. 시계열이 추세를 가진다. (시계열에 증가하거나, 감소하는 추세가 보인다.)
3. 시계열이 계절성을 보인다. 
4. 시계열의 분산이 시간에 따라 변한다. 

--- 

2,3 은 사실 1에 포함되는 개념. 

아래와 같은 자료를 생각해보자. 

```r
e<-rnorm(1000)
t<-1:1000
trend<- (1:1000 / 1000)*30 
## 1:1000 = 1,2,3,...,1000인 벡터가 만들어짐. 
## 1:1000 / 1000 = 1/1000, 2/1000, ... 1인 벡터가 만들어짐. 
## ((1:1000)/1000)*30 = 30/1000, 60/3000, ... 30인 벡터가 만들어짐. 
z<-trend+e
plot(z)
lines(trend,col=2,lwd=3)
```

위의 자료는 단순한 회귀분석으로도 분석가능하다. 즉 위의 자료를 모델로 표현하면 아래와 같다. 

**(모델1)**

$$

---

**따라서 모든 비정상시계열이 분석불가능한것은 아니다!!**

---

분석가능한 비정상시계열이 있다는 것은 알겠음. 

하지만 (모델1)과 같은 시계열은 관심없음. (오차항이 서로 독립이므로, 회귀분석 수준의 분석으로 처리가능함.)

시계열을 배운 의미가 있으려면? 오차항이 서로 독립이 아닌 비정상시계열을 분석해야함. 

예를들면 아래와 같은 자료를 분석할 수 있어야 한다. 

**(모델2) 연습문제 7.6의 모형1**

$$
(Z_t-Z_{t-1})=0.8(Z_{t-1}-Z_{t-2})+\epsilon_t, \quad \epsilon_t \sim N(0,1). 
$$

---

위의 시계열은 분명히 비정상이다. 왜냐하면 위의 자료는 

$$
(1-B)Z_t=0.8(B-B^2)Z_t+\epsilon_t
$$

와 같이 표현할 수 있고,

이는 다시 

$$
(1-B)Z_t=0.8B(1-B)Z_t+\epsilon_t
$$

와 같이 표현가능하고, 이는 다시 


$$
(1-0.8B)(1-B)Z_t=\epsilon_t
$$

와 같이 표현가능하다. 따라서 특성방정식을 풀면 $B=1$을 근으로 가지므로, 이 시계열은 비정상이다. 

---

하지만 $Z_t-Z_{t-1}$ 자체를 새로운 시계열 $W_t$로 생각해보자. 즉 

$$
W_t=Z_t-Z_{t-1}
$$

라고하자. 그러면 

$$
W_t=0.8W_{t-1}+\epsilon_t
$$

와 같이 볼 수 있다. 

--- 

위의 모델은 정상시계열이 된다. 그래서 분석할 수 있다. 

--- 

$Z_t$는 분석할수 없지만, $W_t$는 분석할 수 있다. 

그런데 $W_t$는 $Z_t$를 차분(연속인 경우 미분)하여 얻은 자료이다. 

차분을 했다는 의미는, 
$$
Z_t=\dots,1,2,3,5,\dots
$$
이라면
$$
W_t=\dots,?,1,1,2,\dots
$$
이라는 의미이다. 

즉 $W_t$는 $Z_t$의 변화량, 즉 $\Delta Z_t$로 생각할 수 있다. 

--- 

지금까지의 상황을 요약하면 아래와 같다. 
1. 우리가 관심있는 것은 $Z_t$이다. 
2. 그런데 $Z_t$는 비정상시계열이다. 
3. 그래서 $Z_t$를 분석할 수 없다. 
4. 그런데 $W_t=\Delta Z_t= Z_t - Z_{t-1}$는 정상시계열이 되어 분석할 수 있다. 
---

그렇다면 의문은 "$Z_t$대신에 $W_t$를 분석해도 의미가 있을까?" 이다. 

의미가 있다. 

--- 

아래와 같은 자료를 관찰했다고 가정하자. 
$$
z_{t-1}=1, z_t=2.5
$$

차분을 하면 $w_t=z_t-z_{t-1}=2.5-1=1.5$를 계산할 수 있다. 

아래의 모델을 활용하면, 

$$
W_{t+1}=0.8 W_{t}+\epsilon_{t+1}
$$

$W_{t+1}$을 아래와 같이 추측할 수 있다. 

$$
\hat{W}_{t+1}=0.8\times 1.5 
$$

이때 자료로부터 0.8이라는 숫자는 적절하게 추측하였다고 가정하자. 

그런데 
$$ 
Z_{t+1}=Z_t+W_{t+1}
$$

이므로, $Z_t$자리에 관측한 $z_t=2.5$를 대입하고, $W_{t+1}$대신에 추측한 $0.8\times 1.5$값을 대입하면, 

$$
\hat{Z}_{t+1}=2.5+0.8\times 1.5
$$

와 같이 추측할 수 있다. 

--- 

위의 논의를 일반화 시키면, 1번 차분하여 ARMA(p,q)를 따르는 모델은 모두 분석가능하며 분석의 의미도 있다고 생각할 수 있다. 

이러한 모델을 ARIMA(p,1,q)라고 한다. 차분된 시계열을 $W_t$라고 한다면, ARIMA(p,1,q)는 아래와 같은 모양이 된다. 

$$
{\boldsymbol \phi}(B)W_t={\boldsymbol \theta}(B)\epsilon_t
$$

그런데 $W_t=Z_t-Z_{t-1}=(1-B)Z_t$이므로, 위의 모델은 아래와 같이 쓸 수 있다. 
$$
{\boldsymbol \phi}(B)(1-B)Z_t={\boldsymbol \theta}(B)\epsilon_t
$$

--- 

위의 논의를 다시 일반화 하면 $d$번 차분하여 ARMA(p,q)를 따르는 모델은 모두 분석가능하고 분석의 의미가 있다. 이러한 모델을 ARIMA(p,d,q)라고 하자. ARIMA(p,d,q)의 일반적인 형태는

$$
{\boldsymbol \phi}(B)(1-B)^dZ_t={\boldsymbol \theta}(B)\epsilon_t
$$

와 같이 된다. 

---

예제를 풀어보자. 아래의 모형 

$$
(Z_t-Z_{t-1})=0.8(Z_{t-1}-Z_{t-2})+\epsilon_t, \quad \epsilon_t \sim N(0,1). 
$$

을 정리하면

$$
Z_t=1.8Z_{t-1}-0.8Z_{t-2}+\epsilon_t 
$$

와 같이 된다. 시뮬레이션 해보면,  

```r
set.seed(777)
e<-rnorm(10000)
z<-e*0
z[1]<-0
z[2]<-0
for(t in 3:10000) z[t]<-1.8*z[t-1]-0.8*z[t-2]+e[t] 
plot(z)
```

SACF를 확인할 필요도 없이 비정상시계열처럼 보인다. 

그래도 SACF를 그려보면, 

```r
acf(z)
```

--- 

$Z_t$는 전형적인 비정상 시계열이 된다. 

하지만 차분해서 ARMA를 따르는 경우도 있으므로, 분석을 포기하지 않고 차분을 시도해본다. 

차분은 아래와 같이 하면된다. 

--- 

```r
z[1:5]
z[2]-z[1]
z[3]-z[2]
z[4]-z[3]
z[5]-z[4]
```

---

하지만 $\tt{diff}$함수를 사용해도 된다. 

---

```r
diff(z[1:5])
```

----

원래 $z[1:5]$에는 5개의 자료가 있는데, 차분으로 인하여 1개의 자료가 없어지고 4개의 자료만 남게됨을 주목하라. 

이처럼 1차차분을 하면 1개의 자료가 손상됨을 유의하자. 

동일한 논리로 $d$차차분을 하면 $d$개의 자료가 줄어든다. 

--- 

차분한 자료를 $W_t$라고 두자. 

--- 

```r
w<-c(0,diff(z))
```

---

이때 $W_1=0$으로 가정하여 1개의 손상된 자료를 끼워넣었다. 

$W_t$의 그림을 그려보자. 

---

```r
plot(w)
```

---

정상인것처럼 보인다. 

SACF, SPACF를 그려보자. 


---


```r
acf(w)$acf[1:5]
pacf(w)$acf[1:5]
```
