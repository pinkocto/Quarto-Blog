다음의 모형들에 의해 설명되는 확률과정 $\{Z_t\}$는 정상성을 갖는가? 이를 논하라.


## (c)

$Z_t = A\sin(\frac{2}{3}\pi t + U)$, $A$는 평균이 $0$이고, 분산이 $1$인 확률변수이고 $U$는 상수이다.

## (d)

$Z_t=A\sin(\pi t+U)$, $A$는 평균이 $0$이고 분산이 $1$인 확률변수이고 $U$는 구간 $[-\pi, \pi]$에서 일양분포를 따르는 확률변수이다. 또한 $A$와 $U$는 서로 독립이다.


## (e)

$Z_t = A\cos(\frac{1}{2}\pi t) + B\sin(\frac{1}{2}\pi t)$, $A$와 $B$는 서로 독립이고 각각 평균 $0$, 분산 $1$을 갖는 확률변수들이다.
