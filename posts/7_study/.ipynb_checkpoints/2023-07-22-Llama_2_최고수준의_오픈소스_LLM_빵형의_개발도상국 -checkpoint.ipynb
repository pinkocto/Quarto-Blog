- Llama2-hf : 사용자가 어떤 문장을 입력했을 때 뒤에있는 문장을 이어주는 모델. (Text Completion model)
- Llama2-chat-hf: chatGPT와 같이 chat을 주고받을 수 있는 모델.  (Chat Completion model)

## Load a model

### Llama-2-7b-chat-hf

```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(
    model,
    use_auth_token=True,
)

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

## Helper function

```python
def gen(x, max_length=200):
    sequences = pipeline(
        x,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=max_length,
    )

    return sequences[0]["generated_text"].replace(x, "")
```

```python
print(gen('I liked "Breaking Bad" and "Band of Brothers". Do you have any recommendations of other shows I might like?\n'))
```

```python
print(gen("대한민국에서 유명한 인공지능 유튜버 3명만 나열해봐.", 500))
```

```python
print(gen("유튜브 빵형의 개발도상국 채널 알아?", 500))
```
