## 2. 단위변환 (온도, 길이 등)
> 함수, 포맷팅

````python
def f2c(temp_f):
    return (temp_f - 32) * 5 / 9

def main():
    # 두 줄
    temp_f = 80
    temp_c = f2c(temp_f)
    print(f"{temp_f}F => {temp_c:.2f}C")
    
    # 한 줄
    print(f"{temp_f}F => {f2c(temp_f):.2f}C")
````

````python
temp_f = 80
temp_c = f2c(temp_f)
print(f"{temp_f}F => {temp_c:.2f}C")

# 한 줄
print(f"{temp_f}F => {f2c(temp_f):.2f}C")
````

## 3. 소수 판별하기
> 함수, 조건문(break 등)

````python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

def main():
    num = 8
    
    if is_prime(num):
        print(f"{num}은/는 소수입니다.")
    else:
        print(f"{num}은/는 소수가 아닙니다.")
````

````python
num = 8

if is_prime(num):
    print(f"{num} 은/는 소수입니다.")
else:
    print(f"{num} 은/는 소수가 아닙니다.")
````

## 4. 소수 구하기
> 함수, 반복문

````python
def main():
    list_prime = [x for x in range(1, 1000) if is_prime(x)]
    print(f"1-100까지 중 소수는 {list_prime}입니다.")
    print(f"1-100까지 중 소수의 개수는 {list_prime}입니다.")
````

````python
list_prime = [x for x in range(1, 1000) if is_prime(x)]
print(f"1-100까지 중 소수는 {list_prime}입니다.")
print(f"1-100까지 중 소수의 개수는 {len(list_prime)}입니다.")
````

## 5. 팩토리얼 구하기 
> 재귀함수

````python
# factorial
def fact(num):
    result = 1
    for i in range(1, num+1):
        result = result * i
    return result

def main():
    print(f"{10}!은 {fact(10)}입니다.")
    
if __name__ == "__main__":
    main()
````

````python
def fact(num):
    if num == 1:
        return 1
    return num * fact(num - 1) # 재귀식 이용
````

## 6. 1-100까지 짝수 합

> 반복문, 조건문

`-` 지능형 리스트 (=리스트 컴프리헨션)


````python
def is_even(a):
    return a % 2 == 0

def main():
    evens = [x for x in range(1, 101) if is_even(x)]  # 뒤에 if가 붙으면 filter역할
    sum_even = sum(evens)
    
    print(f"1-100까지 숫자 중 짝수의 합은 {sum_even}입니다.")
````

````python
main()
````
