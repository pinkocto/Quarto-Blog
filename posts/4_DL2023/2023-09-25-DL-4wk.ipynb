## Numerical Computation XII

- Local minimum
    - $u^\top H u$ is positive for all $u$ (positive definite $H$).
- Local maximum
    - $u^\top H u$ is negative for all $u$ (negative definite $H$).
- Saddle point
    - $u^\top H u$ is negative/positive for some $u$.

## Numerical Computation XIII

- Note that the change of gradient w.r.t. each coordinate can be different.
- Newton’s method (second order approximation).


$$f(x) \approx f(x_0) + (x-x_0)^\top \nabla_x f(x_0) + (x-x_0)^\top H(f)(x_0)(x-x_0)$$

$$x^* = x_0 - H(f)(x_0)^{-1}\nabla_x f(x_0)$$

## Numerical Computation XVI

***<font color='red'>4. Constrained Optimization</font>***

$$\arg \min_x f(x) \text{ subject to } x \in \cal{S}.$$

- Here, points that lie within the set $\cal{S}$ are called feasible points.
- Simple approach: projection into $\cal{S}$ when $x = x − \epsilon ∇x f(x).$
- Structured approach: $f(x, y) = f (\theta)$ where $x = \sin(\theta)$, $y = \cos(\theta)$.
Searching minimizing with a constraint $\sqrt{x^2 + y^2}=1$

## Numerical Computation XVII

-  Karush–Kuhn–Tucker(KKT) approach provides a very general solution to constrained optimization

- setup

1. $\cal S = \{\forall_i, g(x)=0, \forall j, h(x) \leq 0 \}$
2. Object: $\min_{x\in {\cal S}} f(x).$
3. Generalized Lagrangian.

$$L(x, \lambda, \alpha) = f(x) + \sum_{i=1}^m \lambda_i g_i(x) + \sum_{j=1}^n \alpha_j h_j(x), \quad \text{  where  } \lambda_i > 0 \text{  and  } \alpha_j \geq 0 $$

## Numerical Computation XVIII

- Goal: $\min_{x\in \cal{S}} f(x) = \min_x \max_{\lambda} \max_{\alpha\geq 0}L(x,\lambda, \alpha).$
