# 5. 패턴파악 및 시각화 

결측치의 패턴? 

## A. `msno.matrix()`

```python
msno.matrix(df)
```

결측치의 숫자는 비슷하지만 C의 경우 결측치가 랜덤하게 발생하였고, A,B 비슷한 패턴 //  D,E 비슷한 패턴으로 결측치가 발생한 것을 볼 수 있다.

## B. `msno.heatmap()`

```python
msno.heatmap(df)
```

결측치의 정보가 A,B와 비슷 // D,E 비슷
