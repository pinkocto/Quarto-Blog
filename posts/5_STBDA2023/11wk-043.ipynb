# 2. Imports 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.tree
import sklearn.ensemble
#---#
import warnings
warnings.filterwarnings('ignore')
#---#
import matplotlib.animation
import IPython
```

# 3. Data + 의사결정나무로 적합

```python
np.random.seed(43052)
temp = pd.read_csv('https://raw.githubusercontent.com/guebin/DV2022/master/posts/temp.csv').iloc[:,3].to_numpy()[:80]
temp.sort()
eps = np.random.randn(80)*3 # 오차
icecream_sales = 20 + temp * 2.5 + eps 
df_train = pd.DataFrame({'temp':temp,'sales':icecream_sales})
df_train
```

```python
# step1 
X = df_train[['temp']]
y = df_train['sales']
# step2 
predictr = sklearn.tree.DecisionTreeRegressor()
# step3
predictr.fit(X,y)
# step4 -- pass
```

```python
plt.plot(X,y,'o')
plt.plot(X,predictr.predict(X),'--')
```

- 전체적으로 오버피팅 느낌.

# 4. 배깅으로 적합 

```python
# step1 
X = df_train[['temp']]
y = df_train['sales']
# step2 
predictr = sklearn.ensemble.BaggingRegressor()
# step3
predictr.fit(X,y)
# step4 -- pass
```

```python
plt.plot(X,y,'o')
plt.plot(X,predictr.predict(X),'--')
```

- 배깅이라는 predictor를 쓰니까 좀 더 스무스 해졌다.

`-` tree 비교 (애니메이션)

```python
fig, ax = plt.subplots(2,1)
plt.close()
#---#
def func(i):
    ax[0].clear()
    sklearn.tree.plot_tree(
        predictr.estimators_[i],
        feature_names=X.columns.tolist(),
        max_depth=1,
        ax=ax[0]
    )
    ax[0].set_title('predictr.estimator')
    #---#
    ax[1].clear()
    my_tree = sklearn.tree.DecisionTreeRegressor()
    my_tree.fit(X_array[samples[i]],y_array[samples[i]])
    sklearn.tree.plot_tree(
        my_tree,
        feature_names=X.columns.tolist(),
        max_depth=1,
        ax=ax[1]
    );
    ax[1].set_title('my_tree')
#---#
ani = matplotlib.animation.FuncAnimation(fig,func,frames=10)
display(IPython.display.HTML(ani.to_jshtml()))
```

## D. 앙상블결과 재현

`-` 최종결과물 (손으로..)

`-` 최종결과물 (코드로 정리)

```python
def ensemble(trees,i=None):
    if i is None:
        i = len(trees) # 전체 트리를 앙상블 (10개)
    else: 
        i = i+1
    yhat = np.stack([tree.predict(X) for tree in trees[:i]]).mean(axis=0) # i번째 트리까지만 앙상블.
    return yhat
```

```python
len(trees) # 지금까는 10개의 트리를 앙상블해서 결과를 냈었음.
```

```python
ensemble(trees)
```

```python
ensemble(trees,0) # 0번트리만 적용
```

```python
ensemble(trees,1) # 0번트리,1번트리의 예측값 평균
```
